# Gaussian Process

```python
import pyro 
import pyro.distributions as dist
import torch
import matplotlib.pyplot as plt
```

```python
#Generate the data
N = 50
X = dist.Uniform(0,5).sample((N,1))
Z = dist.Uniform(0,5).sample((N,1))
y =  0.5*torch.sin(X+Z) + dist.Normal(0,0.1).sample((N,1))

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter3D(X, Z, y, c=y, cmap='Greens')
```
